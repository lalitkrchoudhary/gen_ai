```python
# In pandas, the category data type is used to represent categorical data — data that has a limited number of repeated values (like labels or groups).

# 📌 Simple definition

# A category dtype stores data as:

# unique values (categories) +
# integer codes referencing those values
# 🧠 Example
# import pandas as pd

# data = pd.Series(["Male", "Female", "Male", "Female"])
# print(data.dtype)

# Output:

# object

# Now convert to category:

# data = data.astype("category")
# print(data.dtype)

# Output:

# category
# 🔍 What’s happening internally

# Instead of storing full strings repeatedly:

# ["Male", "Female", "Male"]

# Pandas stores:

# Categories: ["Female", "Male"]
# Codes: [1, 0, 1]
```
